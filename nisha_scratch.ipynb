```python
import requests
from env import github_token, github_username

# Set up API request headers
headers = {
    "Authorization": f"token {github_token}",
    "User-Agent": github_username,
    "Accept": "application/vnd.github.v3+json"
}

def get_repositories(query, page=1):
    api_url = f"https://api.github.com/search/repositories?q={query}&sort=stars&order=desc&per_page=100&page={page}"
    response = requests.get(api_url, headers=headers)
    if response.status_code == 200:
        return response.json()["items"]
    else:
        print(f"Failed to fetch repository list. Status code: {response.status_code}")
        exit()

# Retrieve the top 200 "breast cancer" repositories
repositories_1st_page = get_repositories("breast cancer")
repositories_2nd_page = get_repositories("breast cancer", page=2)

all_repositories = repositories_1st_page + repositories_2nd_page

# Print the results
for repo in all_repositories:
    print(f'/{repo["full_name"]}')

```

```python
total_titles = len(all_repositories)
print(f"Total repository titles fetched: {total_titles}")
```

```python
def get_readme_content(repo_full_name):
    api_url = f"https://api.github.com/repos/{repo_full_name}/readme"
    response = requests.get(api_url, headers=headers)
    if response.status_code == 200:
        # Decoding the content from base64 format
        content_base64 = response.json()["content"]
        decoded_content = base64.b64decode(content_base64).decode('utf-8')
        return decoded_content
    else:
        return None

# Retrieve the top 200 "breast cancer" repositories
repositories_1st_page = get_repositories("breast cancer")
repositories_2nd_page = get_repositories("breast cancer", page=2)

all_repositories = repositories_1st_page + repositories_2nd_page

# Fetching README content for each repository
data_list = []
for repo in all_repositories:
    repo_info = {
        "repo": f'/{repo["full_name"]}',
        "language": repo["language"],
        "readme_contents": get_readme_content(repo["full_name"])
    }
    data_list.append(repo_info)

# Saving data to Github_data.json
with open('Github_data.json', 'w') as f:
    json.dump(data_list, f, indent=4)

print("Data saved to Github_data.json")
```
